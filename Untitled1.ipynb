These lines import the necessary libraries.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

This function performs <strong>gamma correction</strong> on an input image. Gamma correction is a non-linear operation that changes the <strong>brightness of an image</strong>, often used to <strong>correct for monitor or camera discrepancies</strong>. This function takes <strong>two arguments:</strong> <em>the source image and the gamma value</em>. It then calculates the inverse gamma and creates a lookup table to transform the input image based on the gamma value. The transformed image is returned.


```python
# Gamma correction
def gammaCorrection(src, gamma):
    invGamma = 1 / gamma
    table = [((i / 255) ** invGamma) * 255 for i in range(256)]
    table = np.array(table, np.uint8)
    return cv2.LUT(src, table)
```

```python
# Load the image
path = r"C:\Users\yaren\Desktop\2.2.07.tiff"
img = imread(path)
imshow(img); #original image
```

This line applies <strong>gamma correction</strong> to the input image, with a gamma value of <strong>0.15</strong>.



```python
# Gamma correction
gammaImg = gammaCorrection(img, 0.15)
```

These lines display the <strong>gamma-corrected image</strong>.

```python
imshow(gammaImg);
```

This line performs <strong>contrast stretching</strong> on the gamma-corrected image. Contrast stretching is a technique used to <strong>improve the contrast of an image by increasing the dynamic range of the pixel values</strong>. This is achieved by normalizing the pixel values to span the full range of possible values (in this case, 0 to 255).

```python
# Contrast stretching
minmax_img = cv2.normalize(gammaImg, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
```

These lines display the <strong>contrast stretching on the gamma-corrected image</strong>.

```python
imshow(minmax_img);
```

These lines perform <strong>color balancing</strong> on the contrast-stretched image using <strong>histogram equalization</strong>. Histogram equalization is a technique used to adjust the contrast of an image by redistributing the pixel values based on the frequency of occurrence of each pixel value. This is done separately on each color channel of the image. The final color-balanced image is obtained by merging the equalized color channels.

```python
# Color balancing using Histogram Equalization
b, g, r = cv2.split(minmax_img)
equalized_b = cv2.equalizeHist(b)
equalized_g = cv2.equalizeHist(g)
equalized_r = cv2.equalizeHist(r)
equalized_img = cv2.merge((equalized_b, equalized_g, equalized_r))
```

These lines display the <strong>color balancing on the contrast-stretched image</strong> and also this is my <strong>output image</strong>.

```python
imshow(equalized_img);
```
